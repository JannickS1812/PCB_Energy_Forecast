```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [14, 8]
```

```python
df = pd.read_csv(r'data\placer.csv', delimiter=',')

# rename columns
df['recipe'] = df['RECIPENAME']
df['start'] = pd.to_datetime(df['STARTTIME'])
df['end'] = pd.to_datetime(df['ENDTIME'])
df['comp'] = df['NUMCOMP'].astype('int32')
df['block'] = df['NUMBLOCKS'].astype('int32')
df['error'] = df['NUMERRORS'].astype('int32')
df.drop(['Unnamed: 0', 'PCBID', 'RECIPENAME','SIDE','STARTTIME','ENDTIME','MODULENO','NUMCOMP','NUMBLOCKS','NUMERRORS'], axis=1, inplace=True)

df.head()
```

```python
df.plot.scatter(x='block', y='moduleno')
```

```python
df.describe()
```

```python
num_rows = len(df)
```

```python
starts = df.STARTTIME
starts_y = pd.Series([1] * len(starts))

ends = df.ENDTIME
ends_y = pd.Series([1] * len(ends))
```

```python
df_starts = pd.DataFrame(data={'start':starts_y, 'date':pd.to_datetime(df.STARTTIME)})
df_ends = pd.DataFrame(data={'end':ends_y, 'date':pd.to_datetime(df.ENDTIME)})

df_starts.set_index('date')
df_ends.set_index('date')
```

```python
df_furnace = pd.read_csv(r'data\furnace_train_set.csv', delimiter=';')
df_furnace["time"] = pd.to_datetime(df_furnace["time"])
df_furnace["energy"].fillna(0, inplace=True)
df_furnace["energy"] = df_furnace["energy"]/max(df_furnace["energy"])
df_furnace["nitrogen"] = df_furnace["nitrogen_Nm3h"]/max(df_furnace["nitrogen_Nm3h"])

df_eval = pd.read_csv(r'data\task_fill_blanks.csv', delimiter=';')
df_eval["time"] = pd.to_datetime(df_eval["time"])
df_eval["energy"].fillna(0, inplace=True)
df_eval["energy_eval"] = df_eval["energy"]/max(df_eval["energy"])
df_eval["nitrogen_eval"] = df_eval["nitrogen_Nm3h"]/max(df_eval["nitrogen_Nm3h"])
```

```python
ax = df_starts.plot.scatter(x='date', y='start', color='red')
df_ends.plot.scatter(x='date', y='end', ax=ax)
df_furnace.plot(x='time', y='nitrogen', color='black', ax=ax)
df_eval.plot(x='time', y='nitrogen_eval', color='green', ax=ax)
```
